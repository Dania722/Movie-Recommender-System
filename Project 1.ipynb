```python
# series k liye parameter bna rhy hai , to isliye single parameter hi kry gye 
def extract_values (str_lst):
    values = json.loads(str_lst)
    return [i["name"] for i in values]
```

```python
# humy automatic loop chalni hai backend pe  or value return krdy , it can be done through 
## apply or map

# kisis series pe function lgana hai , to map or apply 

# apply only accept single parameter 
movies_credit["genres"] = movies_credit["genres"].apply(extract_values)
```

```python
## mujhy chiye hr row ki cast se staring kii 3 names ly aye , hero herone , villan 

def three (lst):
    return lst[:3]
```

```python

## now we are using as lambda function 
# whi use hota, declare hota return hota and then destroy hojata 

# three_slt is our parameter , save is varialbe mein hoga ,
# : is k bd jo hoga woh return kryga 
movies_credit["cast"] = movies_credit["cast"].apply(extract_values).apply(lambda three_lst:three_lst[:3])
```

```python
# series k liye parameter bna rhy hai , to isliye single parameter hi kry gye 
def fetch_Director (str_lst):
    values = json.loads(str_lst)
    return [i["name"] for i in values if i["job"] == "Director"]
```

```python
def collapes (lst):
    finalLst = []
    for i in lst:
        finalLst.append(i.replace(" " , ""))
    return finalLst
```

```python
def fetch_poster(movie_id):
    url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key=9be5f69f3dce45bfe3a6e88c2cef56fe"
    data = requests.get(url)
    data = data.json()
    poster_path = data['poster_path']
    full_path = "https://image.tmdb.org/t/p/w500/" + poster_path
    return full_path
```

```python
print(fetch_poster(19995))
```

```python
def recommender (m_name):
    m_index = finalData[finalData["title"] == m_name].index[0]
    m_distances = similiarity[m_index]
    index_list = list(enumerate(m_distances))
    similar_movie = sorted(index_list, reverse = True, key = lambda x: x[1])[1:11]
    
    recommended_movie_names = []
    recommended_movie_posters = []
    
    for i in similar_movie:
        # fetch the movie poster
        movie_id = finalData.iloc[i[0]].movie_id
        recommended_movie_posters.append(fetch_poster(movie_id))
        recommended_movie_names.append(finalData.iloc[i[0]].title)

    return recommended_movie_names,recommended_movie_posters
```

```python
recommender("Iron Man 2")
```
